```python
import pytesseract
from PIL import Image
from en import braille_mapping
import re
```

```python
im = Image.open(r"img/2.jpg")
myconfig = r'--psm 6 --oem 3'
txt = pytesseract.image_to_string(image=im, config=myconfig)
```

```python
txt = re.sub(r'(QO\)|Q)\s', '', txt)
txt = re.sub(r'\n+', ' ', txt)
txt
```

```python
for c in txt:
    if (c != ' '):
        map = braille_mapping[c]
        print(f'{c} : {map}')
    else:
        print(f'{c} : {[[0, 0], [0, 0], [0, 0]]}')
```
